```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer la paleta de colores azul-verde
sns.set_palette("viridis")

# Cargar el DataFrame estratificado
stratified_df = pd.read_csv('data_sample_stratified_for_flourish.csv')

# 1. 'percentage_by_employment_status' vs 'labour_force', diferenciado por 'isced11'
plt.figure(figsize=(16, 10))
sns.boxplot(data=stratified_df, x='labour_force', y='percentage_by_employment_status', hue='isced11')
plt.title('Porcentaje por Estado de Empleo según Situación Laboral, diferenciado por Nivel Educativo (ISCED-11)')
plt.xlabel('Situación Laboral')
plt.ylabel('Porcentaje por Estado de Empleo')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Nivel Educativo (ISCED-11)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 2. 'percentage_education_by_birth' vs 'country_birth', diferenciado por 'isced11'
plt.figure(figsize=(20, 12)) # Aumentamos el tamaño para acomodar más países
sns.boxplot(data=stratified_df, x='country_birth', y='percentage_education_by_birth', hue='isced11')
plt.title('Porcentaje de Educación por Nacimiento según País de Nacimiento, diferenciado por Nivel Educativo (ISCED-11)')
plt.xlabel('País de Nacimiento')
plt.ylabel('Porcentaje de Educación por Nacimiento')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Nivel Educativo (ISCED-11)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. 'percentage_education_general' vs 'labour_force', diferenciado por 'isced11'
plt.figure(figsize=(16, 10))
sns.boxplot(data=stratified_df, x='labour_force', y='percentage_education_general', hue='isced11')
plt.title('Porcentaje de Educación General según Situación Laboral, diferenciado por Nivel Educativo (ISCED-11)')
plt.xlabel('Situación Laboral')
plt.ylabel('Porcentaje de Educación General')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Nivel Educativo (ISCED-11)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
